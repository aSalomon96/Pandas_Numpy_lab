#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
## Primero tenemos que pasar la columna duration a int ya que se encuentra en un formato string poco convencional
df_contenido["duration"] = df_contenido["duration"].astype(str) #nos aseguramos que la columna es string

#usamos funcion para limpiar los datos y convertirlos
def limpiar_duracion(valor):
    if isinstance(valor,str): #nos aseguramos que la columna es string
        valor = valor.strip().lower() 
        if 'min' in valor:
            return int(valor.replace(' min', '').strip()) #sacamos el " min" y convertimos a entero
        elif 'season' in valor:  # Manejar valores de "Seasons"
            return None 
        return None
    
#aplicamos funcion
df_contenido["duration"]=df_contenido["duration"].apply(limpiar_duracion)

df_contenido.sample(5)
```

```python
duracion90 = df_contenido[df_contenido.duration==90]
duracion90[["title","duration"]]
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
df_director = df_contenido.query('director=="Christopher Nolan"|director=="Martin Scorsese"')
print(df_director[["title","director"]])

```

```python
# contarr los titulos de cada director
conteo_directores = df_director.groupby('director')['title'].count().reset_index()

# renombrar las columnas para mayor claridad
conteo_directores.columns = ['director', 'cantidad_dirigida']

print(conteo_directores)
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
paises = ['United States', 'United Kingdom', 'Canada'] #creamos la lista de paises que queremos buscar
paises_df = df_contenido[df_contenido["country"].isin(paises)] #usamos isin() para chequear los paises que queremos en el df

#Vemos como son los valores unicos de la columna
paises_df["Language"].unique()
```

```python
patron_regex = "/"#si suponemos que las que tienen mas de dos lengujaes estan separadas por "/"
df_leng = paises_df[paises_df["Language"].str.contains(patron_regex,na=False)] #indicamos con la funcion contains

df_leng[["Language"]].value_counts() #contamos los valores unicos con respecto a los paises el contenido que tiene mas de un lenguaje
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
filtroIMDb = df_contenido[df_contenido["IMDB Score"].between(6.5,8.0,inclusive="both")] #Creamos filtro con between
filtro_duracion = filtroIMDb[filtroIMDb["duration"].isnull()] #filtramos los nulos

filtro_duracion[["title","IMDB Score","duration"]]
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
tv_show= df_contenido[df_contenido.type=="TV Show"] #filtramos primero por tvshow
str_love= "love" #indicamos la palabra clave que queremos filtrar

tv_show_love = tv_show[tv_show["title"].str.contains(str_love, regex = True,case=False)]#creamos un df con la formula contains() que filtra por nuestra palabra clave

tv_show_love[["title"]]
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?
